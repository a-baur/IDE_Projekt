```python
from urllib.parse import quote_plus
import random

def _resource_from_uri(uri):
    if not uri:
        # if uri is missing, create random one
        return f"_{random.randint(0, 99999999):08}"
    return uri.split("/")[-2]


def get_dbr_uri(term: str):
    if not term:
        return
    term = term.replace(' ', '_')
    resource = quote_plus(term)
    return resource
```
